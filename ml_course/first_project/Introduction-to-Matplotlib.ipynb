## Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in Jupyter 
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10, 10)) # (Width, Height)

# 3. Plot data
ax.plot(x, y)

# 4. Customize plot 
ax.set(title="Simple plot",
       xlabel="x-axis",
       ylabel="y-axis")

# 5. Save & show (You save the whole figure)
fig.savefig("images/sample-plot.png")
plt.show()
```

```python
# Make a plot from the dictionary
nut_butter_prices = {"Almond butter": 10, 
                     "Penut butter": 8, 
                     "Cashew butter": 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values()) # (x, y)
ax.set(title= "Cade's Nut Butter Store",
       ylabel="Price ($)")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values())) # barh is a horizontal bar graph
plt.show()
```

```python
# Make some data for histograms and plot it
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
plt.show()
```

### Two options for subplots

```python
# Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(10, 5))
# Plot to each different axis
ax1.plot(x , x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(np.random.randn(1000))

plt.show()
```

```python
# Subplots option 2
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10, 5))
# Plot to each different index
ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.random(10), np.random.random(10))
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1, 1].hist(np.random.randn(1000))

plt.show()
```

### Which method should you use? (pyplot vs matplotlib oo method)

* When plotting something quickly, okay to use the pyplot method
* When plotting something more advanced, use the OO method

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
# Pyplot method
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target",
             figsize=(10, 6))
plt.show()
```

```python
# OO method mixed with pyplot method
fig, ax = plt.subplots(figsize=(10 ,6))
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c= 'target',
             ax=ax);
#ax.set_xlim([45, 100])
plt.show()
```

```python
over_50.target.values
```

```python
# OO method from scratch
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])

# Customize the plot
ax.set(title="Heart Disease and Cholesterol Levels", 
       xlabel="Age",
       ylabel="Cholesterol")

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target")

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),
           linestyle='--')

plt.show()
```

```python
over_50.head()
```

```python
# Subplot of chol, thalach and age
fig, (ax0, ax1) =plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10, 10),
                              sharex= True)
# Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"])
# Customize ax 0
ax0.set(title = "Heart Disease and Cholesterol Levels",
       #xlabel="Age",
       ylabel="Cholesterol")
# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")
# Add a meanline
ax0.axhline(y=over_50["chol"].mean(),
            linestyle="--")

# Add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"])
# Customize ax1
ax1.set(title="Heart Disease and Max Heart Rate Levels",
        xlabel="Age",
        ylabel="Max Heart Rate")
# Add a legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target")
# Add a meanline
ax1.axhline(y=over_50["thalach"].mean(),
            linestyle="--")
# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize = 16, fontweight = "bold")

plt.show()
```

```python
# Histogram
# Subplot of chol, thalach, and age
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10, 10))
# Add data to ax0
hist = ax0.hist(x=over_50["age"], bins=15, label="Age") # color="skyblue",
# Customize ax0
ax0.set(title="Heart Disease and Cholesterol Levels",
        xlabel="Age",
        ylabel="Cholesterol Levels")
# Add a meanline
ax0.axvline(x=over_50["age"].mean(),
            color="red",
            linestyle="--",
            label="Mean Age")
# Add a legend to ax0
ax0.legend()

# Bar graph
# Add data to ax1
bar = ax1.bar(x=over_50["age"],
              height=over_50["thalach"],
              #color="orange",
              label="Max Heart Rate")
# Customize ax1
ax1.set(title="Heart Disease and Max Heart Rate Levels",
        xlabel="Age",
        ylabel="Max Heart Rate")
# Add a meanline
ax1.axhline(y=over_50["thalach"].mean(),
            color="red",
            linestyle="--",
            label="Mean Max Heart Rate")
# Add a legend to ax1
ax1.legend()

# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold")

plt.show()
```

```python
# Set the style
plt.style.use('seaborn-v0_8-whitegrid')

# OO method from scratch
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                     cmap="winter") # this  changes the colour scheme

# Customize the plot
ax.set(title="Heart Disease and Cholesterol Levels", 
       xlabel="Age",
       ylabel="Cholesterol")

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target")

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),
           linestyle='--')

plt.show()
```

```python
# Customizing the y and x axis limitations 

# Subplot of chol, thalach and age
fig, (ax0, ax1) =plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10, 10),
                              sharex= True)
# Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"],
                      cmap="winter")
# Customize ax 0
ax0.set(title = "Heart Disease and Cholesterol Levels",
       #xlabel="Age",
       ylabel="Cholesterol")
# Change ax0 x axis limits
ax0.set_xlim([50, 80])
# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")
# Add a meanline
ax0.axhline(y=over_50["chol"].mean(),
            linestyle="--")

# Add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"],
                      cmap="winter")
# Customize ax1
ax1.set(title="Heart Disease and Max Heart Rate Levels",
        xlabel="Age",
        ylabel="Max Heart Rate")
# Change ax1 x axis limits
ax1.set_xlim([50, 80]) # Not needed because sharex = True
ax1.set_ylim(60, 200)
# Add a legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target")
# Add a meanline
ax1.axhline(y=over_50["thalach"].mean(),
            linestyle="--")
# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize = 16, fontweight = "bold")

plt.show()
```
